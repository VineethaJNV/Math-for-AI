```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 70 # Mean
sigma = 10 # Standard Deviation

x =np.linspace(30, 100, 1000)

y = norm.pdf(x, mu, sigma)

plt.plot(x, y, color='black', linewidth=2, label="normal distribution")

x1 = np.linspace(30, 100, 1000)


y = norm.pdf(x, mu, sigma)

plt.plot(x, y, color="black", linewidth=2, label = "Normal Distribution")

x1 = np.linspace(mu-sigma, mu+sigma, 1000)
plt.fill_between(x1, norm.pdf(x1, mu, sigma), color = "green", alpha = 0.3, label = "68% of  region(1sigma)")

x2 = np.linspace(mu-2*sigma, mu+2*sigma, 1000)
plt.fill_between(x2, norm.pdf(x2, mu, sigma), color = "yellow", alpha = 0.2, label = "95% of  region(2sigma)")

x3 = np.linspace(mu-3*sigma, mu+3*sigma, 1000)
plt.fill_between(x3, norm.pdf(x3, mu, sigma), color = "blue", alpha = 0.1, label = "68% of  region(3sigma)")

plt.title("Normal Distribution of student scores")
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```
